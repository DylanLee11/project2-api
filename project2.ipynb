## Part 2: Explore one Wikipedia page with the beautifulsoup package

```python
import bs4
import requests
import pandas as pd
```

```python
#save and print the text content of a page with all tags removed

url = 'https://en.wikipedia.org/wiki/Chicago_Bears_Stadium'
response = requests.get(url)
soup = bs4.BeautifulSoup(response.text, 'html.parser')
text_content = soup.get_text()
print(text_content)
```

```python
#download an image with beautifulsoup and save it in this repository
image_url = 'https://upload.wikimedia.org/wikipedia/en/thumb/4/4a/New_Chicago_Bears_Stadium.jpg/375px-New_Chicago_Bears_Stadium.jpg'
image_response = requests.get(image_url)
with open('ChicagoBearsStadium.jpg', 'wb') as file:
    file.write(image_response.content)
```

```python
#find all the links in a page with beautifulsoup
#print the first 100 characters of ten of these links
url = 'https://en.wikipedia.org/wiki/Chicago_Bears_Stadium'
page = requests.get(url)
soup = bs4.BeautifulSoup(page.content, 'html.parser')

links = soup.find_all('a', href=True)

for i, link in enumerate(links[:10], 1):
    link_href = str(link.get('href'))[:100]
    print(f"Link {i}: {link_href}")
```

```python
#using the links in the "link" column, download the first 1000 characters of each script
#use requests and bs4, remember to remove all html tags
def download_script_text(url):
    try:
        response = requests.get(url)
        soup = bs4.BeautifulSoup(response.text, 'html.parser')
        text = soup.get_text()
        return text[:1000]
    except:
        return None

scripts['script_text'] = scripts['link'].apply(download_script_text)
```

## Part 4: TMDB database

#### Browse the documentation at https://developer.themoviedb.org/reference/intro/getting-started. Create an account to authenticate

```python
#create a dataset of the movies in theaters now. Include metadata fields you are interested in. 
import requests
import pandas as pd
import os

# Set up TMDB API
api_key = 'c5bc01cc3a53e1d54a06399626d24a42'
base_url = 'https://api.themoviedb.org/3'
endpoint = f'{base_url}/movie/now_playing'
params = {
    'api_key': api_key,
    'language': 'en-US',
    'page': 1
}

# Fetch movies in theaters
response = requests.get(endpoint, params=params)
movies = response.json().get('results', [])

# Extract metadata fields
movie_data = []
for movie in movies:
    movie_info = {
        'title': movie.get('title'),
        'release_date': movie.get('release_date'),
        'overview': movie.get('overview'),
        'vote_average': movie.get('vote_average'),
    }
    movie_data.append(movie_info)

# Create a DataFrame
df = pd.DataFrame(movie_data)

# Save the DataFrame to a CSV file
df.to_csv('movies_in_theaters.csv', index=False)
```

```python
#download the movie posters for 10 of these movies and save them to this repository
poster_base_url = 'https://image.tmdb.org/t/p/w500'
os.makedirs('posters', exist_ok=True)

for movie in movie_data[:10]:
    poster_path = movie.get('poster_path')
    if poster_path:
        poster_url = f'{poster_base_url}{poster_path}'
        poster_response = requests.get(poster_url)
        if poster_response.status_code == 200:
            with open(f"posters/{movie['title'].replace('/', '_')}.jpg", 'wb') as f:
                f.write(poster_response.content)

# Display the DataFrame
df
```
